I'm using the June 2025 Digital City Map from [here](https://www.nyc.gov/content/planning/pages/resources/datasets/digital-city-map)

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the shapefile
df = gpd.read_file('data/DCM_StreetCenterLine.shp')
df.head()
```

```python
df['Street_NM'] = df['Street_NM'].str.lower()
df['words_count'] = df['Street_NM'].str.count(" ")+1
df.sort_values('words_count', ascending=False).head(10)
```

#### The below list of street types is based on my own knowledge of roads and was supplemented by some research (and just zooming around on Google Maps to make sure I'm not missing any obvious road types)

```python
street_types = ['avenue', 'boulevard', 'bridge', 'court', 'concourse', 'drive', 'expressway', 'highway', 'lane', 'parkway', 'plaza', 'road', 'row', 'street', 'terrace', 'turnpike', 'way']
```

#### Then I defined a simple function that uses a list comprehension to return all the matching words in the street names then applied the function to my `Street_NM` column

```python
def street_types_analysis (words):
    matches = [word for word in words if word in street_types]
    return " ".join(matches)
```

```python
df['ind_street_type'] = df['Street_NM'].str.split().apply(street_types_analysis)
df.head()
```

```python
df['street_type_count'] = df['ind_street_type'].str.count(' ')+1
```
