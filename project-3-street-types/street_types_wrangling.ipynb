I'm using the June 2025 Digital City Map from [here](https://www.nyc.gov/content/planning/pages/resources/datasets/digital-city-map)

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Read the shapefile
df = gpd.read_file('data/DCM_StreetCenterLine.shp')
df.head()
```

```python
df = df[df['Feat_statu']=="City_St"]
```

```python
df['Street_NM'] = df['Street_NM'].str.lower()
df['words_count'] = df['Street_NM'].str.count(" ")
df.sort_values('words_count', ascending=True).head(50)
```
